__SLCA's Weighted Average Cost of Capital (WACC)__
<br>
$ WACC = \frac{E}{V} \times R_e + \frac{D}{V} \times R_d \times (1 - T_c) $
<br>
Where:
- $ E $ is the market value of equity
- $ V $ is the total market value of equity and debt
- $ R_e $ is the cost of equity
- $ D $ is the market value of debt
- $ R_d $ is the cost of debt
- $ T_c $ is the tax rate

```python
# Calculate the market value of equity
market_value_equity = hist_data['Close'] * yf.Ticker(slca).info['sharesOutstanding']
market_value_equity
```
