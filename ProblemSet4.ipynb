Peers Analysis

```python
peers = {'RPC Inc': 'RES', 'ProPetro Holding Corp': 'PUMP', 'Cactus': 'WHD', 'Core Laboratories': 'CLB', 'ChampionX Corp':'CHX', 'Patterson-UTI Energy': 'PTEN', 'Liberty Energy Inc.': 'LBRT'}
```

```python
# Download data for the peers
peers_data = yf.download(list(peers.values()), start='2019-01-01', end='2024-04-26')
```

```python
peers_data.head()
```

```python
# Get the income statement, balance sheet, and cash flow statement for the peers
income_statement_peers = {}
balance_sheet_peers = {}
cash_flow_peers = {}

for peer in peers.values():
    peer_data = yf.Ticker(peer)
    income_statement_peers[peer] = peer_data.financials
    balance_sheet_peers[peer] = peer_data.balance_sheet
    cash_flow_peers[peer] = peer_data.cashflow
```

```python
# Plot the close price of the peers and slca for the past three years
plt.figure(figsize=(12, 8))
plt.plot(hist_data['Close'][-756:], label='SLCA Close Price')
for peer in peers.values():
    plt.plot(peers_data['Close'][peer][-756:], label=f'{peer} Close Price')
plt.title('Close Price of SLCA and Peers (3 Yrs)')
plt.legend(loc='best')
plt.show()
```

```python
# Plot the peers and slca for the last year
plt.figure(figsize=(12, 8))
plt.plot(hist_data['Close'][-252:], label='SLCA Close Price')
for peer in peers.values():
    plt.plot(peers_data['Close'][peer][-252:], label=f'{peer} Close Price')
plt.title('Close Price of SLCA and Peers (1 Yr)')
plt.legend(loc='best')
plt.show()
```

Financial Analysis of SLCA

```python
# Find the market cap of SLCA for the past five years
market_cap_slca = hist_data['Close'] * yf.Ticker(slca).info['sharesOutstanding']
market_cap_slca.head()
```

```python
# Create a function that transposes the financial statements
def transpose_financials(financials):
    return financials.T
```

```python
# Transpose the income statement, balance sheet, and cash flow statement
income_statement_t = transpose_financials(income_statement)
balance_sheet_t = transpose_financials(balance_sheet)
cash_flow_t = transpose_financials(cash_flow)
```

```python
income_statement_t.columns
```

```python
# Profitability Ratios: gross profit margin, operating margin, net profit margin, return on assets, and return on equity
# Gross Profit Margin
gross_profit_margin = (income_statement_t['Total Revenue'] - income_statement_t['Cost Of Revenue']) / (income_statement_t['Total Revenue']) * 100
# Rename the column
gross_profit_margin = gross_profit_margin.rename('Gross Profit Margin')
gross_profit_margin
```

```python
# Operating Margin EBIT / Total Revenue
operating_margin = income_statement_t['EBIT'] / income_statement_t['Total Revenue'] * 100
# Rename the column
operating_margin = operating_margin.rename('Operating Margin')
operating_margin
```

```python
# Net Profit Margin Net Income / Total Revenue
net_profit_margin = income_statement_t['Net Income'] / income_statement_t['Total Revenue'] * 100
# Rename the column
net_profit_margin = net_profit_margin.rename('Net Profit Margin')
net_profit_margin
```

```python
# Return on Assets Net Income / Total Assets
return_on_assets = income_statement_t['Net Income'] / balance_sheet_t['Total Assets'] * 100
# Rename the column
return_on_assets = return_on_assets.rename('Return on Assets')
return_on_assets
```

```python
# Return on Equity Net Income / Total Equity
return_on_equity = income_statement_t['Net Income'] / balance_sheet_t['Stockholders Equity'] * 100
# Rename the column
return_on_equity = return_on_equity.rename('Return on Equity')
return_on_equity
```

```python
# Liquidity Ratios: current ratio, quick ratio, and cash ratio
# Current Ratio Current Assets / Current Liabilities
current_ratio = balance_sheet_t['Current Assets'] / balance_sheet_t['Current Liabilities']
# Rename the column
current_ratio = current_ratio.rename('Current Ratio')
current_ratio
```

```python
# Quick Ratio (Current Assets - Inventory) / Current Liabilities
quick_ratio = (balance_sheet_t['Current Assets'] - balance_sheet_t['Inventory']) / balance_sheet_t['Current Liabilities']
# Rename the column
quick_ratio = quick_ratio.rename('Quick Ratio')
quick_ratio
```

```python
# Cash Ratio Cash / Current Liabilities
cash_ratio = balance_sheet_t['Cash And Cash Equivalents'] / balance_sheet_t['Current Liabilities']
# Rename the column
cash_ratio = cash_ratio.rename('Cash Ratio')
cash_ratio
```

```python
# Inventory Turnover Cost of Goods Sold / Average Inventory
inventory_turnover = income_statement_t['Cost Of Revenue'] / ((balance_sheet_t['Inventory'] + balance_sheet_t['Inventory']) / 2)
# Rename the column
inventory_turnover = inventory_turnover.rename('Inventory Turnover')
inventory_turnover
```

```python
# Receivables Turnover Total Revenue / Average Accounts Receivable
receivables_turnover = income_statement_t['Total Revenue'] / ((balance_sheet_t['Accounts Receivable'] + balance_sheet_t['Accounts Receivable']) / 2)
# Rename the column
receivables_turnover = receivables_turnover.rename('Receivables Turnover')
receivables_turnover
```

```python
# Leverage Ratios
# Debt to Equity Ratio Total Liabilities / Total Equity
debt_to_equity = balance_sheet_t['Total Liabilities Net Minority Interest'] / balance_sheet_t['Stockholders Equity']
# Rename the column
debt_to_equity = debt_to_equity.rename('Debt to Equity Ratio')
debt_to_equity
```

```python
# Debt to Assets Ratio Total Liabilities / Total Assets
debt_to_assets = balance_sheet_t['Total Liabilities Net Minority Interest'] / balance_sheet_t['Total Assets']
# Rename the column
debt_to_assets = debt_to_assets.rename('Debt to Assets Ratio')
debt_to_assets
```

```python
# Interest Coverage Ratio EBIT / Interest Expense
interest_coverage = income_statement_t['EBIT'] / income_statement_t['Interest Expense']
# Rename the column
interest_coverage = interest_coverage.rename('Interest Coverage Ratio')
interest_coverage
```

```python
# Check to see if the ratios have a tz-aware index or not
for ratio in [gross_profit_margin, operating_margin, net_profit_margin, return_on_assets, return_on_equity, current_ratio, quick_ratio, cash_ratio, inventory_turnover, receivables_turnover, debt_to_equity, debt_to_assets, interest_coverage, price_to_earnings, price_to_book]:
    print(ratio.index.tz)

# Convert the date index from tz-naive to tz-aware
for ratio in [gross_profit_margin, operating_margin, net_profit_margin, return_on_assets, return_on_equity, current_ratio, quick_ratio, cash_ratio, inventory_turnover, receivables_turnover, debt_to_equity, debt_to_assets, interest_coverage, price_to_earnings, price_to_book]:
    ratio.index = ratio.index.tz_convert('America/New_York')
```
